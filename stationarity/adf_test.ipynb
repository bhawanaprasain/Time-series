# Augmented Dickey-Fuller test

Augmented Dickey-Fuller test test is based on a hypothesis and can tell us the degree of probability to which it can be accepted. It is often classified under one of the unit root tests. The test begins with the null and alternate hypotheses.

Ho (Null Hypothesis): The time series data is non-stationary
<br>
H1 (alternate Hypothesis): The time series data is stationary
<br>

Let us assume alpha = 0.05 which means 95% confidence.
<br>
The test results are interpreted using  p-value. If p > 0.05 accept the null hypothesis, else if p <= 0.05 reject the null hypothesis.

```python
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.stattools import adfuller
```

```python
def auto_regressive_process(lags, coefs, length):
  
    coefs = np.array(coefs)
    time_series = [np.random.normal() for lag in range(lags)]
    for _ in range(length):
        #previous values of the series, reversed
        prev_vals = time_series[-lags:][::-1]
        
        #new value of time series
        new_val = np.sum(np.array(prev_vals) * coefs) + np.random.normal()
        time_series.append(new_val)
        
    return np.array(time_series)
```

```python
def adf_test(series):
    result = adfuller(series)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
```


## Auto-regressive process with lag 1

Let us generate data for AR series with lag 1, which is stationary as unit root does not equal to 1.

```python
ar_1_process = auto_regressive_process(1, [.3], 100)
plt.figure(figsize=(10,4))
plt.plot(ar_1_process)
plt.title('Stationary AR process with lag 1', fontsize=18)
```

```python
adf_test(ar_1_process)
```

Since p-value <=  0.5, we reject null hypothesis . So the data is stationary

## Non stationarity

Let us generate data for AR series with lag 1, which is not stationary as unit root equals to 1.

```python

ar_1_process_unit_root = auto_regressive_process(1, [1], 100)
plt.figure(figsize=(10,4))
plt.plot(ar_1_process_unit_root)
plt.title('Non-Stationary AR process with lag 1', fontsize=18)
```

```python
adf_test(ar_1_process_unit_root)
```

Since p-value >  0.5, we accept null hypothesis . So the data is not stationary
